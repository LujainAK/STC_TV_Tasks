```python
df = dataframe.copy()
```

```python
# Check that it copied the dataframe correctly
df
```

```python
# Build the model
y = df['Total_watch_time_in_houres']
model = sm.tsa.SARIMAX(y, order=(0, 1, 1), seasonal_order=(0, 1, 1, 12), enforce_stationarity=True, enforce_invertibility=False)
result = model.fit(disp=0)
```

```python
# Make future predictions
forecast = result.get_forecast(steps=60)
```

```python
# Get the forecasted values
forecast_values = forecast.predicted_mean
```

```python
# Create a DataFrame to display the forecast results
forecast_df = pd.DataFrame({'Total_watch_time_in_houres': forecast_values})
forecast_df = forecast_df.rename_axis("date_").reset_index()
forecast_df.set_index('date_', inplace=True)

# Print the forecast DataFrame
forecast_df
```

```python
# show the predicted dataframe
fig = px.line(forecast_df,  y="Total_watch_time_in_houres")
fig.update_traces(line_color="red")
fig.update_layout(font=dict(size=20))
fig.show()
```
